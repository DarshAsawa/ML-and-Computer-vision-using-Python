```python
#remove null or nan from age.
def lw(cols):
    age = cols[0]
    Pclass = cols[1]
    if pd.isnull(age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        elif Pclass == 3:
            return 25
        else:
            return 30
    else:
        return age
```

```python
dataset['Age'] = dataset[['Age', 'Pclass']].apply(lw , axis=1)
age = dataset['Age']
```

```python
sns.heatmap(dataset.isnull())  
#you can see there is no null value in age
```

```python
#Age is categorical variable, 
#string : number(label encoding) : dummy variable : one hot encoding.

sex = dataset['Sex']
```

```python
sex = pd.get_dummies(sex, drop_first=True)
```

```python
#even pclass is a numerical categorical variable 
pclass = dataset['Pclass']  
```

```python
pclass = pd.get_dummies(pclass,drop_first = True) 
```

```python
#sibblings
sibsp = dataset['SibSp']
sibsp = pd.get_dummies(sibsp, drop_first=True)
```

```python
#parent child
parch = dataset['Parch']
parch = pd.get_dummies(parch, drop_first=True)
```

```python
embarked = dataset['Embarked']
embarked = pd.get_dummies(embarked,drop_first = True)
```
